
# API to CSV

### Import libraries

```python
import urllib
import requests
import json
import csv
from datetime import datetime
```

### Import the API data
* In the API I searched for 'python AND coding'
* Sorted it by oldest to newest in the API 
* Also made sure the field 'wordcount' was used


```python
URL = "https://content.guardianapis.com/search?order-by=oldest&show-fields=wordcount&page-size=50&q=python%20AND%20coding&api-key=58d8c960-8869-4795-8e1a-68d37d318707"
response = urllib.request.urlopen(URL)
print(response)
```

### Load to json
* As the json is nested and only keys that are needed for our desired output are 'response' and 'keys'
* data is printed to make sure the right data is extracted


```python
data = json.loads(response.read())['response']['results']
print(data)
```

### Saving to csv 
* Firstly the csv name is stated as 'output.csv'
* Delare the row names 
* Exlude any articles with a wordcount less than 1000
* Format the web publication date to ddmmyyyy in new column
* Extract just the year to new coloumn 


```python
with open("data-output.csv", 'w') as file:
    csv_file = csv.writer(file)
    csv_file.writerow(['id','type','sectionId', 'sectionName', 'webPublicationDate','formatted_date','year', 'webTitle', 
                      'webUrl', 'apiUrl', 'isHosted', 'pillarId','PillarName', 'wordcount'])
    for item in data:
        if int(item['fields']['wordcount']) >= 1000:
            publishDate = datetime.strptime(item['webPublicationDate'], "%Y-%m-%dT%H:%M:%SZ")
            formatted_date = str(publishDate.day) + "/" + str(publishDate.month) + "/" + str(publishDate.year)
            year = str(publishDate.year)
            csv_file.writerow([item['id'], item['type'], item['sectionId'],
                               item['sectionName'], item['webPublicationDate'],formatted_date,year,
                            item['webTitle'], item['webUrl'], item['apiUrl'], item['isHosted'],
                               item['pillarId'], item['pillarName'], item['fields']['wordcount']])
```
